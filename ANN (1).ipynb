```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
df_train = pd.read_csv("/content/train.csv")
```

```python
df_test = pd.read_csv("/content/test.csv")
```

```python
df_val = pd.read_csv("/content/val.csv")
```

```python
df_train_cleaned = df_train.dropna(subset=['label'])
x_train = df_train_cleaned.drop(["label","id"], axis=1).values/255.0
y_train = df_train_cleaned["label"].values
```

```python
df_val_cleaned = df_val.dropna(subset=['label'])
x_val = df_val_cleaned.drop(["label","id"], axis=1).values/255.0
y_val = df_val_cleaned["label"].values
```

```python
x_test = df_test.drop(["id"],axis = 1).values/255.0
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_val = sc.transform(x_val)
```

```python
y_train_labels = df_train_cleaned["label"].values
y_val_labels = df_val_cleaned["label"].values
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train_labels.astype(int), num_classes=10)
y_val = to_categorical(y_val_labels.astype(int), num_classes=10)
```

```python
from tensorflow.keras.utils import to_categorical
ct = to_categorical
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam,SGD,RMSprop
```

```python
model50 = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation='sigmoid', input_shape=(784,)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model100 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model200 = tf.keras.Sequential([
    tf.keras.layers.Dense(200, activation='sigmoid', input_shape=(784,)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model300 = tf.keras.Sequential([
    tf.keras.layers.Dense(300, activation='sigmoid', input_shape=(784,)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model50.compile(optimizer = Adam(learning_rate = 0.001),loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
history1 = model50.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=30,
    batch_size=20
)
```

```python
history2 = model100.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=30,
    batch_size=20
)
```

```python
models = [history1, history2, history3, history4]

for i, h in enumerate(models):
    print(f"Model {i+1}")
    print("Final Train Accuracy:", h.history["accuracy"][-1])
    print("Final Val Accuracy:", h.history["val_accuracy"][-1])
    print("Final Val Loss:", h.history["val_loss"][-1])
    print("--------------")
```
